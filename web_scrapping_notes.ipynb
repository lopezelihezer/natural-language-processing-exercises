```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
response = requests.get('https://web-scraping-demo.zgulde.net/news')
```

```python
def parse_news_articles(article):
    title = article.h2.text
    date, author = article.select('.italic')[0].find_all('p')
    return {'title': title, 'date':date.text, 'author':author.text, 'text': }
```

```python
soup = BeautifulSoup(response.text)
articles = soup.select('.grid.gap-y-12 > div')
pd.DataFrame([parse_news_articles(article) for article in articles])
```
